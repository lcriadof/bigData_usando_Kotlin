Con describe() obtenemos un detalle de todo el dataframe:

Por un lado:

 **count**: el número de valores no nulos en cada columna.
 
 **min**: el valor mínimo de cada columna.
 
 **max**: el valor máximo de cada columna.
 
 **unique**: el número de valores únicos de cada columna categórica, es decir, el número de valores diferentes que hay en la columna.
  
 **nulls**: el número de valores nulos de cada columna, es decir, el número de valores que no tienen ningún dato.
 
 
Pero también nos facilita medidas de dispersión estadísticas como: 
 
 **mean**: la media aritmética de cada columna.
 
 **std**: la desviación estándar de cada columna.
 
 **median**: la mediana de cada columna numérica, es decir, el valor que divide la distribución en dos partes iguales.
 
 **top**: es la moda de cada columna categórica, es decir, el valor que más se repite. 
 
 **freq**: es la frecuencia absoluta de la moda de cada columna categórica, es decir, el número de veces que se repite el valor más común. 


```kotlin
// si queremos, por ejemplo ver solo la std
raw_df.describe()["std"]
```

```kotlin
// nos podemos quedar solo con las columnas numéricas, facilmente, ademas en la operación limiamos los nulls asignando al año 1966
var df_num=raw_df["Rank", "Year", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]
    // imputing
    .fillNulls { Year }.with { 1966 }
    
df_num.describe()
```

```kotlin
// filtramos entre los años 2006 y 2008
raw_df.filter { Year in 2006..2008 }
```

```kotlin
// contamos registros agrupado por años
var df_years=df_num.groupBy { Year }.count()
// los ordenamos
df_years.sortBy(df_years.Year)
```

```kotlin
// Nos quedamos con unos pocos años
var df=raw_df.filter { Year in 2016..2018 }

var df_years2=df.groupBy { Year }.count()
println(df_years2)
```
