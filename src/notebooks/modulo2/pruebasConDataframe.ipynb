## Formas de crear DATAFRAMEs

```kotlin
%use dataframe(0.9.1)
// dataframe es similar a pandas de python
```

```kotlin
// creamos un dataframe a partir de una LISTA INMUTABLE

data class Person(val name: String, val age: Int)
val persons = listOf(Person("Alice", 15), Person("Bob", 20), Person("Charlie", 22))
// Convert collection to DataFrame
val df_ListaInmutable = persons.toDataFrame()

// visualizamos
df_ListaInmutable
```

```kotlin
// creamos un dataframe a partir de una LISTA MUTABLE

data class Persona(var nombre: String)
var nombres: MutableList<Persona> = mutableListOf()
nombres.add(Persona("Luis"))
nombres.add(Persona("Alfonso"))
nombres.add(Persona("Alberto"))

// Convert collection to DataFrame
var df_Lista_Mutable = nombres.toDataFrame()

// visualizamos
df_Lista_Mutable
```

```kotlin
// directamente
val df = dataFrameOf("firstName", "lastName", "age", "city", "weight", "isHappy")(
        "Alice", "Cooper", 102, "Paris", 54, true,
        "Bob", "Dylan", 45, "Dubai", 87, true,
        "Charlie", "Daniels", 20, "Moscow", null, false,
        "Charlie", "Chaplin", 110, "Milan", null, true,
        "Bob", "Marley", 30, "Tokyo", 68, true,
        "Alice", "Wolf", 20, null, 55, false,
        "Charlie", "Byrd", 105, "Moscow", 90, true
    ).group("firstName", "lastName").into("name")
```

```kotlin
// a partir de columnas

// create columns
val fromTo by columnOf("LoNDon_paris", "MAdrid_miLAN", "londON_StockhOlm", "Budapest_PaRis", "Brussels_londOn")
val flightNumber by columnOf(10045.0, Double.NaN, 10065.0, Double.NaN, 10085.0)
val recentDelays by columnOf("23,47", null, "24, 43, 87", "13", "67, 32")
val airline by columnOf("KLM(!)", "{Air France} (12)", "(British Airways. )", "12. Air France", "'Swiss Air'")

// create dataframe
val df_desde_columnas = dataFrameOf(fromTo, flightNumber, recentDelays, airline)

df_desde_columnas
```
